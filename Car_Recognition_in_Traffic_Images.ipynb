```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

```python
img = 'https://www.its.ucla.edu/wp-content/uploads/sites/6/2018/01/los-angeles-1396606_960_720.jpg'
```

```python
results = model(img)
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```
